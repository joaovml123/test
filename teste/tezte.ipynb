```python
import psycopg2
from psycopg2 import OperationalError

def create_connection():
    try:
        # Define connection parameters
        connection = psycopg2.connect(
            database="postgres",
            user="postgres",
            password="a1s2d3f4",
            host="192.168.15.88",
            port="5432"
        )
        print("Connection to PostgreSQL DB successful")
        return connection
    except OperationalError as e:
        print(f"The error '{e}' occurred")
        return None

def get_daily_sums(connection):
    try:
        cursor = connection.cursor()
        query = """
        SELECT "data"::date AS day, SUM(valor) AS total_valor
        FROM public.venda
        GROUP BY "data"::date
        ORDER BY day;
        """
        cursor.execute(query)
        results = cursor.fetchall()
        for row in results:
            print(f"Day: {row[0]}, Total Value: {row[1]}")
    except OperationalError as e:
        print(f"The error '{e}' occurred")

# Create the connection
conn = create_connection()

# Get daily sums if connection was successful
if conn:
    get_daily_sums(conn)
    conn.close()
```
